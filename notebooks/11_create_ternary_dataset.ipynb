```python
import sys 
from pathlib import Path 
sys.path.append(str(Path().resolve().parents[0]))

import pandas as pd

from config import PROCESSED_DATA_DIR, BERT_DIR
```

```python
reviews_df = pd.read_parquet(PROCESSED_DATA_DIR / "amazon_reviews_features.parquet")
```

```python
reviews_df = reviews_df.drop_duplicates().reset_index(drop=True)
```

```python
print(reviews_df.columns)
```

```python
reviews_df = reviews_df[["rating", "full_text"]]
print(reviews_df.columns)
reviews_df = reviews_df.drop_duplicates().reset_index(drop=True)
```

```python
reviews_df["rating"] = reviews_df["rating"].map({1: 0, 2: 0, 3: 1, 4: 2, 5: 2})
reviews_df = reviews_df.drop_duplicates().reset_index(drop=True)
print(reviews_df["rating"].value_counts().sort_index())
```

```python
# Verify no duplicates before saving to disk
print(reviews_df.duplicated().sum())
```

```python
reviews_df.to_parquet(BERT_DIR / "amazon_reviews_bert_ternary.parquet", index=False)
```
